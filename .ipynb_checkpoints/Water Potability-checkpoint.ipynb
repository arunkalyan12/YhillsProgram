# Spliting of data

```python
x = df.drop(['Potability'], axis=1)
y=df['Potability']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
model=LogisticRegression(solver='liblinear')
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
dtree = DecisionTreeClassifier(criterion='gini')
dtree.fit(x_train,y_train)
```

```python
dtree.score(x_train,y_train)
```

```python
dtree.score(x_test,y_test)
```

# Making predictions

## Cecision Tree Classifier

```python
dTreeR = DecisionTreeClassifier(criterion ='gini',max_depth = 3, random_state = 0)
dTreeR.fit(x_train, y_train)
print(dTreeR.score(x_train, y_train))
```

```python
y_predict = dTreeR.predict(x_test)
```

```python
print(dTreeR.score(x_test,y_test))
```

### Confusion Martix with decision tree classifier

```python
from sklearn import metrics
```

```python
cm = metrics.confusion_matrix(y_test, y_predict, labels=[0,1])

df_cm = pd.DataFrame(cm, index = [i for i in ['No', 'Yes']], columns = [i for i in [['No', 'Yes']]])

plt.figure(figsize = (7,5))
sns.heatmap(df_cm, annot = True, fmt = 'g')
```

## Confusion Martix with adaboost classifie

```python
from sklearn.ensemble import AdaBoostClassifier
abcl = AdaBoostClassifier(n_estimators = 120, random_state=0)
abcl = abcl.fit(x_train,y_train)
y_predict = abcl.predict(x_test)
print(abcl.score(x_test,y_test))
```

```python
cm = metrics.confusion_matrix(y_test, y_predict, labels=[0,1])

df_cm = pd.DataFrame(cm, index = [i for i in ['No', 'Yes']], columns = [i for i in [['No', 'Yes']]])

plt.figure(figsize = (7,5))
sns.heatmap(df_cm, annot = True, fmt = 'g')
```

## Bagging classifier

```python
from sklearn.ensemble import BaggingClassifier
bgcl = BaggingClassifier(n_estimators = 100,
    max_samples=1.0,
    max_features=1.0,
    bootstrap=True,
    bootstrap_features=True,
    oob_score=False,
    warm_start=False,
    n_jobs=None,
    random_state=None,
    verbose=0,)
bgcl = bgcl.fit(x_train,y_train)
y_predict = bgcl.predict(x_test)
print(bgcl.score(x_train,y_train))
print(bgcl.score(x_test,y_test))
```

## Gradient boosting classifier

```python
from sklearn.ensemble import GradientBoostingClassifier
gbcl = GradientBoostingClassifier()
gbcl = gbcl.fit(x_train,y_train)
y_predict = gbcl.predict(x_test)
print(gbcl.score(x_train,y_train))
print(gbcl.score(x_test,y_test))
```

## Random forest classifier

```python
from sklearn.ensemble import RandomForestClassifier
rfcl = RandomForestClassifier( max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    min_weight_fraction_leaf=0.0,
    max_features='auto',
    max_leaf_nodes=None,
    min_impurity_decrease=0.0,
    n_estimators=10,
    criterion='gini',                          
    bootstrap=True,
    oob_score=False,
    n_jobs=None,
    random_state=None,
    verbose=0,
    warm_start=False,
    class_weight=None,
    ccp_alpha=0.0,
    max_samples=None,)
rfcl = rfcl.fit(x_train,y_train)
y_predict = rfcl.predict(x_test)
print(rfcl.score(x_train,y_train))
print(rfcl.score(x_test,y_test))
```

## SVC

```python
from sklearn.svm import SVC
svcl = SVC()
svcl = svcl.fit(x_train, y_train)
y_predict = svcl.predict(x_test)
print(svcl.score(x_train,y_train))
print(svcl.score(x_test,y_test))
```
